```python
cal_per_id = data[['Id','Calories','TotalDistance','VeryActiveMinutes']]
```

```python
cal_per_id = cal_per_id.groupby('Id')[['Calories','TotalDistance','VeryActiveMinutes']].mean().round(2)
```

```python
cal_per_id.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotting a 3D scatter plot
ax.scatter(cal_per_id['TotalDistance'], cal_per_id['Calories'], cal_per_id['VeryActiveMinutes'], c='blue', marker='o')

ax.set_xlabel('Total Distance')
ax.set_ylabel('Calories')
ax.set_zlabel('Very Active Minutes')

plt.title('Relationship between Calories, Total Distance, and Very Active Minutes')
plt.show()
```

```python
sleep_day = pd.read_csv(r'/content/drive/MyDrive/FitBit Dataset/sleepDay_merged.csv')
```

```python
sleep_day.dtypes
```

```python
sleep_day['SleepDay'] = pd.to_datetime(sleep_day['SleepDay']).dt.strftime('%d, %b, %y')
```

```python
sleep_day['hourAsleep'] = sleep_day['TotalMinutesAsleep']/60
sleep_day['TotalTimeInBed'] = sleep_day['TotalTimeInBed']/60
```

```python
sleep_day.head()
```

```python
sleep_day_mean = sleep_day[['TotalSleepRecords', 'hourAsleep', 'TotalTimeInBed']].mean().sort_values().round(1)
sleep_day_mean
```

```python
ax = sleep_day_mean.plot(kind='bar', color=['yellow', 'grey', 'pink'])
plt.title('Mean Values for Sleep Metrics')
plt.xlabel('Sleep Metrics')
plt.ylabel('Mean Value')
plt.xticks(rotation=90)

ax.bar_label(ax.containers[0], labels=[f'{label}: {value:.2f}' for label, value in sleep_day_mean.items()], color='black', fontsize=8, label_type='edge')

plt.legend()
plt.show()
```
